```python
import random
numero= random.randint(1,101)
tentativa=0
pontuacao=500
print("Bem vindo ao jogo de ADVINHAÇÃO!\n")
print('********************************')
print('\nAcabei de pensar em um número, tenta advinhar!')
print('\n')
print('Em qual nível que desejar jogar?')
print('(1) - Facil \n(2) - Médio \n(3) - Difícil.\n')
print('Lembrando que a qualquer momento para sair do jogo digite 0.')
nivel=int(input('Nível do jogo: '))
while nivel<1 or nivel>3:
  print('Por favor digite 1, 2 ou 3')
  nivel=int(input('Nível do jogo: '))
print('\n')

if nivel ==1:
  tentativa=20
  print("Bem vindo ao nível Fácil do jogo. Se prepare para advinhar.\n")
elif nivel==2:
  tentativa=10
  print("Bem vindo ao nível Médio do jogo!\n")
elif nivel ==3:
  tentativa=5
  print('Bem vindo ao nível Difícil.\n')
print('____________________________________\n')
for jogo in range(tentativa):
  palpite=int(input("Digite um número: "))
  
  if palpite <0 or palpite >100:
    print('Por favor digite numero entre 1 e 100')
    continue
  if palpite==numero:
      print('----------------------------------------')
      print('\n******* PARABENS, Voce acertou *******')
      break
  elif palpite==0:
      print('Você saiu do jogo')
      break
      
  elif palpite < numero:
      print(f'O numero que pensei é maior que {palpite}.')
      rodada=tentativa-1-jogo
      
      if rodada==0:
        print(f'Pensei no número {numero} ')
        print('\n******* Você Perdeu *********')
       
        
      else:
        print(f'Voce tem mais {rodada} chances\n')
             
  elif palpite> numero:
      print(f'O numero que pensei é menor que o {palpite}')
      rodada=tentativa-1-jogo
      
      if rodada==0:
        print(f'Pensei no número {numero} ')
        print('\n\n******* Você Perdeu *********')
        
      else:
        print(f'Voce tem mais {rodada} chances\n')
  
  
  chute=abs(numero-palpite)
  pontuacao=pontuacao-chute
print(f'\nPensei no número {numero}')
print(f'Sua pontução é {pontuacao}')
```
